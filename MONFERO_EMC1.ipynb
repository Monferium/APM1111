```python
from itertools import combinations
import pandas as pd

# Define the population
population = ['Marcos', 'Marcos', 'Marcos', 'Marcos', 'Robredo', 'Robredo']

# Generate all possible combinations of 4 elements from the population
all_samples = list(combinations(population, 4))

# Function to compute proportions of each name in a sample
def compute_proportions(sample):
    count_robredo = sample.count('Robredo')
    total = len(sample)
    return {
        'Robredo': count_robredo / total,
        'Marcos': (total - count_robredo) / total
    }

# Compute proportions for all samples
proportions = [compute_proportions(sample) for sample in all_samples]

# Create a DataFrame for the samples
df_samples = pd.DataFrame(all_samples, columns=['Element 1', 'Element 2', 'Element 3', 'Element 4'])
df_proportions = pd.DataFrame(proportions)

# Combine the samples and proportions DataFrames
df_combined = pd.concat([df_samples, df_proportions], axis=1)

# Compute average proportions
average_proportions = df_proportions.mean()

# Display average proportions
print("\nAverage proportions:")
print(f"Average proportion of 'Robredo': {average_proportions['Robredo']:.2f}")
print(f"Average proportion of 'Marcos': {average_proportions['Marcos']:.2f}")

# Optional: Save the DataFrame to a CSV file for further analysis
df_combined.to_csv('sample_proportions.csv', index=False)
```
